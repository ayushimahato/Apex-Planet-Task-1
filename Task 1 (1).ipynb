```python
# HOTEL BOOKING CANCELLATION DATA CLEANING PROJECT
# Intern Name: Your Name
# Objective:
# Clean and prepare hotel booking data for analysis

import pandas as pd
import numpy as np

df = pd.read_csv("hotel_bookings.csv")
df.head()
```

```python
print("Rows and Columns:", df.shape)
df.info()
df.describe()
```

# Missing Value Analysis

```python
df.isnull().sum()
```

# Handle missing values

```python
df['agent'] = df['agent'].fillna(0)
df['country'] = df['country'].fillna(df['country'].mode()[0])
df['children'] = df['children'].fillna(df['children'].median())
print(df.isnull().sum())

```

```python
print('company' in df.columns)
```

```python
df.drop('company', axis=1, inplace=True, errors='ignore')
```

```python
print('company' in df.columns)
```

# Duplicate Analysis

```python
duplicates = df.duplicated().sum()

print("No. of duplicate values found:", duplicates)
```

```python
# Remove duplicates
duplicates = df.duplicated().sum()
print(f"Number of duplicate records found: {duplicates}")

df = df.drop_duplicates()
```

```python
print(f"Remaining duplicates: {df.duplicated().sum()}")
```

```python
print(df.shape)
print(df.duplicated().sum())
```

```python
df[['adults','children','babies']].describe()
```

# Remove Unrealistic Values

```python
rows_before = len(df)

df = df[
    (df['adults'] >= 1) &
    (df['adults'] <= 10)
]
df = df[df['children'] <= 5]
df = df[df['babies'] <= 5]

rows_after = len(df)

print(f"Rows before outlier removal: {rows_before}")
print(f"Rows after outlier removal: {rows_after}")
print(f"Rows removed: {rows_before - rows_after}")
print(df[['adults','children','babies']].max())
```

# Handle Agent Missing Values

```python
print("Missing values in agent:", df['agent'].isnull().sum())

df['agent'] = df['agent'].fillna(0)

print("Missing values after cleaning:", df['agent'].isnull().sum())


print("Missing values in country:", df['country'].isnull().sum())

df['country'] = df['country'].fillna(df['country'].mode()[0])

print("Missing values after cleaning:", df['country'].isnull().sum())


print("Missing values in children:", df['children'].isnull().sum())

df['children'] = df['children'].fillna(df['children'].median())

print("Missing values after cleaning:", df['children'].isnull().sum())


print(df.isnull().sum())
```

# Outlier Detection

```python
numerical_cols = ['lead_time', 'adr', 'adults', 'children', 'babies']

for col in numerical_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1

    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]

    print(f"{col}: {len(outliers)} outliers found")
```

# Verify New Columns

```python
# Total Guests
df['total_guests'] = (
    df['adults'] +
    df['children'] +
    df['babies']
)
# Total nights
df['total_nights'] = (
    df['stays_in_weekend_nights'] +
    df['stays_in_week_nights']
)
# Booking Status
df['booking_status'] = df['is_canceled'].map({
    0: 'Completed',
    1: 'Canceled'
})

print(df[['total_guests', 'total_nights', 'booking_status']].head())
```

```python
# Arrival Date Column
df['arrival_date'] = pd.to_datetime(
    df['arrival_date_day_of_month'].astype(str)
    + '-' +
    df['arrival_date_month']
    + '-' +
    df['arrival_date_year'].astype(str)
)

print(df['arrival_date'].head())
```

```python
df.to_csv("cleaned_hotel_bookings.csv", index=False)

print("Cleaned dataset saved successfully!")
```
